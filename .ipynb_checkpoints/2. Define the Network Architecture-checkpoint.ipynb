## PyTorch Neural Nets

To define a neural network in PyTorch, you define the layers of a model in the function `__init__` and define the feedforward behavior of a network that employs those initialized layers in the function `forward`, which takes in an input image tensor, `x`. The structure of this Net class is shown below and left for you to fill in.

Note: During training, PyTorch will be able to perform backpropagation by keeping track of the network's feedforward behavior and using autograd to calculate the update to the weights in the network.

#### Define the Layers in ` __init__`
As a reminder, a conv/pool layer may be defined like this (in `__init__`):
```
# 1 input image channel (for grayscale images), 32 output channels/feature maps, 3x3 square convolution kernel
self.conv1 = nn.Conv2d(1, 32, 3)

# maxpool that uses a square window of kernel_size=2, stride=2
self.pool = nn.MaxPool2d(2, 2)      
```

#### Refer to Layers in `forward`
Then referred to in the `forward` function like this, in which the conv1 layer has a ReLu activation applied to it before maxpooling is applied:
```
x = self.pool(F.relu(self.conv1(x)))
```

Best practice is to place any layers whose weights will change during the training process in `__init__` and refer to them in the `forward` function; any layers or functions that always behave in the same way, such as a pre-defined activation function, should appear *only* in the `forward` function.

## Apply the model on a test sample

To test the model on a test sample of data, you have to follow these steps:
1. Extract the image and ground truth keypoints from a sample
2. Make sure the image is a FloatTensor, which the model expects.
3. Forward pass the image through the net to get the predicted, output keypoints.

This function test how the network performs on the first batch of test data. It returns the images, the transformed images, the predicted keypoints (produced by the model), and the ground truth keypoints.

````python
# test the model on a batch of test images

def net_sample_output():

    # iterate through the test dataset
    for i, sample in enumerate(test_loader):

        # get sample data: images and ground truth keypoints
        images = sample['image']
        key_pts = sample['keypoints']

        # convert images to FloatTensors
        images = images.type(torch.FloatTensor)

        # forward pass to get net output
        output_pts = net(images)

        # reshape to batch_size x 68 x 2 pts
        output_pts = output_pts.view(output_pts.size()[0], 68, -1)

        # break after first image is tested
        if i == 0:
            return images, output_pts, key_pts
````

## Visualize the predicted keypoints

Once we've had the model produce some predicted output keypoints, we can visualize these points in a way that's similar to how we've displayed this data before, only this time, we have to "un-transform" the image/keypoint data to display it.

Note that I've defined a *new* function, `show_all_keypoints` that displays a grayscale image, its predicted keypoints and its ground truth keypoints (if provided).

````python
def show_all_keypoints(image, predicted_key_pts, gt_pts=None):
    """Show image with predicted keypoints"""
    # image is grayscale
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=20, marker='.', c='m')
    # plot ground truth points as green pts
    if gt_pts is not None:
        plt.scatter(gt_pts[:, 0], gt_pts[:, 1], s=20, marker='.', c='g')
````

#### Un-transformation

Next, you'll see a helper function. `visualize_output` that takes in a batch of images, predicted keypoints, and ground truth keypoints and displays a set of those images and their true/predicted keypoints.

This function's main role is to take batches of image and keypoint data (the input and output of your CNN), and transform them into numpy images and un-normalized keypoints (x, y) for normal display. The un-transformation process turns keypoints and images into numpy arrays from Tensors *and* it undoes the keypoint normalization done in the Normalize() transform; it's assumed that you applied these transformations when you loaded your test data.

````python
# visualize the output
# by default this shows a batch of 10 images
def visualize_output(test_images, test_outputs, gt_pts=None, batch_size=10):

    for i in range(batch_size):
        plt.figure(figsize=(20,10))
        ax = plt.subplot(1, batch_size, i+1)

        # un-transform the image data
        image = test_images[i].data   # get the image from it's wrapper
        image = image.numpy()   # convert to numpy array from a Tensor
        image = np.transpose(image, (1, 2, 0))   # transpose to go from torch to numpy image

        # un-transform the predicted key_pts data
        predicted_key_pts = test_outputs[i].data
        predicted_key_pts = predicted_key_pts.numpy()
        # undo normalization of keypoints
        predicted_key_pts = predicted_key_pts*50.0+100

        # plot ground truth points for comparison, if they exist
        ground_truth_pts = None
        if gt_pts is not None:
            ground_truth_pts = gt_pts[i]
            ground_truth_pts = ground_truth_pts*50.0+100

        # call show_all_keypoints
        show_all_keypoints(np.squeeze(image), predicted_key_pts, ground_truth_pts)

        plt.axis('off')

    plt.show()

# call it
visualize_output(test_images, test_outputs, gt_pts)
````

````python
def train_net(n_epochs, train_loader, val_loader=None):
    # prepare the net for training
    net.train()

    train_losses_history = [] # List to store average training loss per epoch
    val_losses_history = []   # List to store average validation loss per epoch

    for epoch in range(n_epochs):  # loop over the dataset multiple times

        # Training Phase
        running_loss_batch_print = 0.0 # For printing every 10 batches
        running_epoch_train_loss_sum = 0.0  # To calculate average epoch loss
        total_train_samples = 0       # To count total samples processed in training epoch

        for batch_i, data in enumerate(train_loader):
            images = data['image']
            key_pts = data['keypoints']

            # flatten pts
            key_pts = key_pts.view(key_pts.size(0), -1)

            # convert variables to floats for regression loss
            key_pts = key_pts.type(torch.FloatTensor)
            images = images.type(torch.FloatTensor)

            # zero the parameter (weight) gradients
            optimizer.zero_grad()

            # forward pass to get outputs
            output_pts = net(images)

            # calculate the loss between predicted and target keypoints
            loss = criterion(output_pts, key_pts)

            # backward pass to calculate the weight gradients
            loss.backward()

            # update the weights
            optimizer.step()

            # print loss statistics
            running_loss_batch_print += loss.item()
            running_epoch_train_loss_sum += loss.item() * images.size(0) # Accumulate sum of losses for average epoch loss
            total_train_samples += images.size(0)

            if batch_i % 10 == 9:    # print every 10 batches
                print('Epoch: {}, Batch: {}, Avg. Batch Loss: {:.4f}'.format(epoch + 1, batch_i+1, running_loss_batch_print/10))
                running_loss_batch_print = 0.0

        avg_epoch_train_loss = running_epoch_train_loss_sum / total_train_samples
        train_losses_history.append(avg_epoch_train_loss)
        wandb.log({"train_loss": avg_epoch_train_loss}, step=epoch) # Log training loss to W&B

        # Validation Phase (if val_loader is provided)
        if val_loader:
            net.eval() # Set model to evaluation mode
            running_epoch_val_loss_sum = 0.0
            total_val_samples = 0
            with torch.no_grad(): # Disable gradient calculations for validation
                for val_batch_i, val_data in enumerate(val_loader):
                    val_images = val_data['image']
                    val_key_pts = val_data['keypoints']

                    val_key_pts = val_key_pts.view(val_key_pts.size(0), -1)
                    val_key_pts = val_key_pts.type(torch.FloatTensor)
                    val_images = val_images.type(torch.FloatTensor)

                    val_output_pts = net(val_images)
                    val_loss = criterion(val_output_pts, val_key_pts)

                    running_epoch_val_loss_sum += val_loss.item() * val_images.size(0)
                    total_val_samples += val_images.size(0)

            avg_epoch_val_loss = running_epoch_val_loss_sum / total_val_samples
            val_losses_history.append(avg_epoch_val_loss)
            wandb.log({"val_loss": avg_epoch_val_loss}, step=epoch) # Log validation loss to W&B
            net.train() # Set model back to training mode for next epoch

            print(f"Epoch: {epoch+1}/{n_epochs}, Train Loss: {avg_epoch_train_loss:.4f}, Validation Loss: {avg_epoch_val_loss:.4f}")
        else:
            print(f"Epoch: {epoch+1}/{n_epochs}, Train Loss: {avg_epoch_train_loss:.4f}")


    print('Finished Training')
    return train_losses_history, val_losses_history # Return both lists

# Train your network using the configured number of epochs
n_epochs = wandb.config.epochs # Get epochs from wandb.config

# Call train_net with both train_loader and test_loader (as validation)
train_losses, val_losses = train_net(n_epochs, train_loader, test_loader)

print("\nFinal Training Losses per Epoch:", train_losses)
print("Final Validation Losses per Epoch:", val_losses)
````

````python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Assuming test_images, test_outputs, and gt_pts are available from previous execution
# If not, uncomment the line below to re-run net_sample_output()
# test_images, test_outputs, gt_pts = net_sample_output()

# Ensure both are on the CPU and converted to numpy for easier handling if needed
predicted_keypoints = test_outputs.detach().cpu().numpy()
ground_truth_keypoints = gt_pts.cpu().numpy()

# Calculate squared differences for each coordinate (x, y)
squared_diff = (predicted_keypoints - ground_truth_keypoints)**2

# Sum squared differences across x and y for each keypoint
# Resulting shape: (batch_size, 68)
sum_squared_diff_per_keypoint = np.sum(squared_diff, axis=2)

# Calculate Mean Squared Error (MSE) for each keypoint across the batch
# Resulting shape: (68,)
mse_per_keypoint = np.mean(sum_squared_diff_per_keypoint, axis=0)

# Calculate Root Mean Squared Error (RMSE) for each keypoint
rmse_per_keypoint = np.sqrt(mse_per_keypoint)

print(f"RMSE for each of the 68 keypoints:\n{rmse_per_keypoint}")
print(f"Average RMSE across all keypoints: {np.mean(rmse_per_keypoint):.4f}")

# Plotting the RMSE for each keypoint
plt.figure(figsize=(15, 6))
plt.bar(range(len(rmse_per_keypoint)), rmse_per_keypoint, color='skyblue')
plt.xlabel('Keypoint Index')
plt.ylabel('RMSE (Normalized)')
plt.title('RMSE for Individual Keypoints (Normalized Coordinates)')
plt.xticks(range(0, len(rmse_per_keypoint), 5)) # Show fewer x-ticks for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

This plot shows the RMSE for each of the 68 facial keypoints, calculated using the normalized keypoint coordinates. A lower bar indicates better accuracy for that specific keypoint. This can help identify which facial features are more challenging for the model to predict accurately.

### Question 1: What optimization and loss functions did you choose and why?


Loss Function (Criterion): I chose nn.MSELoss() (Mean Squared Error). This is a common and appropriate choice for regression tasks like facial keypoint detection. It measures the average of the squares of the errors between the predicted keypoint coordinates and the true keypoint coordinates, penalizing larger errors more significantly. Our goal is to predict continuous (x, y) coordinates for each keypoint, making MSE a suitable metric to minimize.
Optimizer: I selected optim.Adam. Adam is an adaptive learning rate optimization algorithm that's widely used and generally performs well across a variety of deep learning tasks. It computes adaptive learning rates for each parameter, which often leads to faster convergence and more stable training compared to simpler optimizers like SGD, especially in complex models.

### Question 2: What kind of network architecture did you start with and how did it change as you tried different architectures? Did you decide to add more convolutional layers or any layers to avoid overfitting the data?

Multiple Convolutional Layers (convX_Y) and Max-Pooling Layers (poolX): These layers are designed to extract increasingly complex features from the input images, such as edges, textures, and ultimately facial features.
Adaptive Average Pooling (AdaptiveAvgPool2d): This layer helps in making the network more robust to input size variations, although our input is fixed at 224x224 after transformations.
Fully-Connected Layers (fcX): These layers take the high-level features learned by the convolutional layers and map them to the final output (136 keypoint coordinates).
Dropout Layer (Dropout(p=0.5)): A dropout layer with a probability of 0.5 was included. Dropout is a regularization technique used to prevent overfitting by randomly setting a fraction of input units to 0 at each update during training. This prevents co-adaptation of neurons and forces the network to learn more robust features.

### Question 3: How did you decide on the number of epochs and batch_size to train your model?

Number of Epochs (n_epochs): The initial suggestion in the notebook was to start with a small number of epochs (e.g., 1 or 2) to get a quick initial observation of the model's training behavior. Following this, I increased n_epochs to 10. This decision was based on observing the training loss decreasing steadily over the initial epochs, indicating that the model was learning. Training for 10 epochs allows for a more substantial period for the model to learn and converge without necessarily overfitting, given the relatively small number of epochs. In a real-world scenario, this number would be determined by monitoring validation loss and stopping when it ceases to improve.
Batch Size (batch_size): The batch size was set to 10. This is a relatively small batch size. A smaller batch size can introduce more noise into the gradient updates, which can sometimes help the model escape shallow local minima and generalize better. However, it also means more frequent updates and potentially slower overall training time per epoch. For this project, a batch size of 10 was chosen as a reasonable starting point to balance training speed and learning effectiveness, as it allows for frequent weight updates.

## Feature maps

Each CNN has at least one convolutional layer that is composed of stacked filters (also known as convolutional kernels). As a CNN trains, it learns what weights to include in it's convolutional kernels and when these kernels are applied to some input image, they produce a set of **feature maps**. So, feature maps are just sets of filtered images; they are the images produced by applying a convolutional kernel to an input image. These maps show us the features that the different layers of the neural network learn to extract. For example, you might imagine a convolutional kernel that detects the vertical edges of a face or another one that detects the corners of eyes. You can see what kind of features each of these kernels detects by applying them to an image. One such example is shown below; from the way it brings out the lines in an the image, you might characterize this as an edge detection filter.
